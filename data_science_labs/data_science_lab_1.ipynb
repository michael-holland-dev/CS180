## Exercise 2: Pointclouds
A pointcloud is a 3D data structure that defines a set of points in 3D space. In autonomous driving, vehicles are outfitted with Lidar ("light detection and ranging”), which are scanning devices that use lasers to measure the distance between a source and nearby objects. Their primary output is a point cloud representing a 3D snapshot of the world around them. Typically point clouds are stored in object geometry files called .obj files. This is essentially a text file where each line represents a different data structure (i.e., points, faces, textures, etc…). See the [Wavefront.obj](https://en.wikipedia.org/wiki/Wavefront_.obj_file#File_format) wiki page for more details.

image-20210810-230840.png

### Note:

In some cases the face may contain multiple values separate by a forward slash /. In this case you can parse the list of ints and simply take whatever appears before the first slash. 

For example, in line 21, the face is defined as f 5/1 4/1 3/1. The resulting list would be [5, 4, 3]. 

### Example:

Lets take a concrete look at an example:
